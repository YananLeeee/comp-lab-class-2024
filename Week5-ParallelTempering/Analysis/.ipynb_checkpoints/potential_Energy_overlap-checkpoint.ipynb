```python
import numpy as np
import matplotlib.pyplot as plt

# Load energy data from each replica's .xvg file
# Assuming you have formatted data in columns [Time, Potential Energy]
replica1_data = np.loadtxt('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/setup/T300/potential_energy.xvg', comments=['@', '#'])
replica2_data = np.loadtxt('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/setup/T400/potential_energy.xvg', comments=['@', '#'])
replica3_data = np.loadtxt('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/setup/T600/potential_energy.xvg', comments=['@', '#'])

# Extract potential energy columns
energy_replica1 = replica1_data[:, 1]
energy_replica2 = replica2_data[:, 1]
energy_replica3 = replica3_data[:, 1]

# Create histograms
bins = 50  # You can adjust the number of bins for finer/coarser histograms

plt.hist(energy_replica1, bins=bins, alpha=0.5, label='Replica 1', density=True)
plt.hist(energy_replica2, bins=bins, alpha=0.5, label='Replica 2', density=True)
plt.hist(energy_replica3, bins=bins, alpha=0.5, label='Replica 3', density=True)

# Add labels and title
plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Probability Density')
plt.title('Energy Histogram Overlap')
plt.legend()

# Show the plot
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/Figures/potential_Energy_overlap.png')
plt.show()
```
