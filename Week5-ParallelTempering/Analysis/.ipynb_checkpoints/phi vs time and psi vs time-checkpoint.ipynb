```python
import mdtraj as md
import matplotlib.pyplot as plt
# Load the trajectory and structure
traj = md.load('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/setup/alanine dipeptide/md_alanine.trr', top='/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/setup/alanine dipeptide/md_alanine.gro')
# Compute phi and psi angles
phi_indices, phis = md.compute_phi(traj)
psi_indices, psis = md.compute_psi(traj)
# Get the time in nanoseconds
time = traj.time / 1000  # Convert from picoseconds to nanoseconds

# Plot phi angles over time
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time, phis[:, 0], label='Phi', color='blue')  # Use first phi dihedral
plt.xlabel('Time (ns)', fontsize=22)
plt.ylabel('Phi (radians)', fontsize=22)
plt.title('Phi vs Time', fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.grid(True)

# Plot psi angles over time
plt.subplot(2, 1, 2)
plt.plot(time, psis[:, 0], label='Psi', color='orange')  # Use first psi dihedral
plt.xlabel('Time (ns)', fontsize=22)
plt.ylabel('Psi (radians)', fontsize=22)
plt.title('Psi vs Time', fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.grid(True)

plt.tight_layout()
# Save the figure to the Figures folder
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/Figures/phi_psi_vs_time.png')
plt.show()

```
