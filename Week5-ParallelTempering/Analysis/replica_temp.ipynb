```python
import numpy as np
import matplotlib.pyplot as plt

#  replica_temp.xvg 
def read_xvg(filename):
    """
    .xvg
    """
    data = np.loadtxt(filename, comments=['#', '@'])
    return data

#  replica_temp.xvg 
replica_temp_data = read_xvg('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/setup/T300/replica_temp.xvg')

# 100
frame_indices = np.arange(0, len(replica_temp_data), 100)
downsampled_data = replica_temp_data[frame_indices]

# 
frames = downsampled_data[:, 0]  # 
temperatures = downsampled_data[:, 1:]  # 

# 
plt.figure(figsize=(10, 6))
for i in range(temperatures.shape[1]):
    plt.plot(frames, temperatures[:, i], label=f'Replica {i+1}')

plt.xlabel('Frames')
plt.ylabel('Temperature (K)')
plt.title('Replica Temperature Movement Over Time')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week5-ParallelTempering/Figures/replica_temp.png')
plt.show()
```
