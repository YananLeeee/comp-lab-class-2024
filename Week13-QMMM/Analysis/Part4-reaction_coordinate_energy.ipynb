```python
import numpy as np
import matplotlib.pyplot as plt

# File paths
file_path = "/home/yl12451/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog"
output_path = "/home/yl12451/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Figures/metad1_initmonitor_Reaction_Coordinate_vs_Time.png"

# Metadynamics parameters from the provided .inp file
hill_height = 1.5  # kcal/mol
hill_width = 0.5   # Å
hill_frequency = 100  # steps

try:
    # Load data
    data = np.loadtxt(file_path)
    time = data[:, 0]  # First column: time (fs)
    reaction_coordinate = data[:, 1]  # Second column: reaction coordinate

    # Convert time to ps
    time_ps = time / 1000  # Convert fs to ps

    # Plot Reaction Coordinate vs. Time
    plt.figure(figsize=(10, 6))
    plt.plot(time_ps, reaction_coordinate, label=f"Metadynamics\n(Hill Height: {hill_height} kcal/mol, "
                                                 f"Hill Width: {hill_width} Å, "
                                                 f"Hill Frequency: {hill_frequency} steps)",
             color="orange", lw=2)

    # Annotate the reaction coordinate formula
    reaction_coordinate_formula = r"RC = d1 - d2"
    plt.text(0.9, 0.9, f"Formula: {reaction_coordinate_formula}", transform=plt.gca().transAxes,
             fontsize=12, horizontalalignment='right', bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white"))

    # Add labels, title, and grid
    plt.xlabel("Time (ps)", fontsize=12)
    plt.ylabel("Reaction Coordinate (Å)", fontsize=12)
    plt.title("Reaction Coordinate vs. Time (Metadynamics)", fontsize=14)
    plt.grid(True)

    # Add legend
    plt.legend(fontsize=10, loc='upper left')

    # Save the plot
    plt.tight_layout()
    plt.savefig(output_path)
    plt.show()

    print(f"Plot saved at: {output_path}")

except FileNotFoundError:
    print(f"Error: File not found at {file_path}. Please check the file path and try again.")
except IndexError:
    print("Error: The data file does not have the expected number of columns. Check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
