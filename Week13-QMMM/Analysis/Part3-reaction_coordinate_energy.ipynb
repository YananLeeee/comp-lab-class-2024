```python
import numpy as np
import matplotlib.pyplot as plt

# File paths
file_path = "/home/yl12451/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Outputs/monitor/MONITOR-COLVAR.metadynLog"
output_path = "/home/yl12451/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Figures/monitor_Reaction_Coordinate_vs_Time.png"

try:
    # Load data
    data = np.loadtxt(file_path)
    time = data[:, 0]  # First column: time (fs)
    reaction_coordinate = data[:, 1]  # Second column: reaction coordinate

    # Convert time to ps
    time_ps = time / 1000  # Convert fs to ps

    # Plot Reaction Coordinate vs. Time
    plt.figure(figsize=(8, 5))
    plt.plot(time_ps, reaction_coordinate)

    # Annotate the reaction coordinate formula
    reaction_coordinate_formula = r"RC = d1 - d2"
    plt.text(0.05, 0.95, f"Formula: {reaction_coordinate_formula}", transform=plt.gca().transAxes,
             fontsize=10, verticalalignment='top', bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white"))

    # Add labels, title, and grid
    plt.xlabel("Time (ps)", fontsize=12)
    plt.ylabel("Reaction Coordinate ()", fontsize=12)
    plt.title("Reaction Coordinate vs. Time", fontsize=14)
    plt.grid(True)


    # Add legend
    plt.legend(fontsize=10)

    # Save the plot
    plt.tight_layout()
    plt.savefig(output_path)
    plt.show()

    print(f"Plot saved at: {output_path}")

except FileNotFoundError:
    print(f"Error: File not found at {file_path}. Please check the file path and try again.")
except IndexError:
    print("Error: The data file does not have the expected number of columns. Check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
