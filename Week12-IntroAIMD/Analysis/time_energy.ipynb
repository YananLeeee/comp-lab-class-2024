```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("/home/yl12451/comp-lab-class/comp-lab-class-2024/Week12-IntroAIMD/Runs/Water/water_aimd_1ns_every0.5fs-1.ener")  # Replace with your file
time = data[:, 0]  # Assuming time is in the first column
energy = data[:, 1]  # Assuming energy is in the second column

plt.plot(time, energy)
plt.xlabel("Time (fs)")
plt.ylabel("Energy (eV)")
plt.title("Energy vs. Time")
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week12-IntroAIMD/Figures/water.png')
plt.show()
```
