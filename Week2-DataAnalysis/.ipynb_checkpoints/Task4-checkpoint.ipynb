```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Set up the plot style to a default style
plt.style.use('ggplot')  # You can use 'ggplot', 'bmh', 'fivethirtyeight', or 'classic'

# File path to the data file
file_path = './1hz3_T310.run.25000000.energy.xvg'

def read_xvg(file_path):
    """
    Read and parse an .xvg file, skipping metadata and returning data as a DataFrame.
    """
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Initialize lists to hold data
    time = []
    data = []

    # Process each line
    for line in lines:
        if line.startswith(('@', '#')):
            continue  # Skip comment lines
        # Split line by whitespace and convert to floats
        values = list(map(float, line.split()))
        if len(values) > 1:  # Check if line has data
            time.append(values[0])
            data.append(values[1:])

    # Convert lists to DataFrame
    df = pd.DataFrame(data)
    df.insert(0, 'Time', time)  # Insert time column at the beginning
    return df

# Read the data file
try:
    data = read_xvg(file_path)
except FileNotFoundError:
    print(f"File not found: {file_path}")
    raise

# Check the first few rows to understand the structure of the data
print(data.head())

# Extract the columns
time_ns = data['Time'] / 1000  # Convert time from ps to ns
quantities = data.iloc[:, 1:]

# Define the quantity names based on typical formats in energy files
quantity_names = [
    'Potential Energy',
    'Kinetic Energy',
    'Total Energy',
    'Temperature',
    'Pressure',
    'Density'
]

# Plot each quantity vs time
fig, axes = plt.subplots(3, 2, figsize=(15, 12), sharex=True)
axes = axes.flatten()

for i, (col, name) in enumerate(zip(quantities.columns, quantity_names)):
    axes[i].plot(time_ns, quantities[col])
    axes[i].set_title(name)
    axes[i].set_xlabel('Time (ns)')
    axes[i].set_ylabel(name)

plt.tight_layout()
plt.show()

# Plot Kinetic Energy and Temperature on the same plot with two y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:red'
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(time_ns, quantities[3], color=color, label='Temperature')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Kinetic Energy', color=color)  # we already handled the x-label with ax1
ax2.plot(time_ns, quantities[1], color=color, label='Kinetic Energy')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Kinetic Energy and Temperature vs Time')
plt.show()
```
