```python
import numpy as np
import matplotlib.pyplot as plt

# Load the grid data
data = np.loadtxt('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Setup/Lugano_exercise_1/A/B10_S0.1_H1.2_grid.dat', comments='#')

# Assuming the energy column is the third column (adjust as needed)
energy = data[:, 2]  # Adjust index based on your grid file structure
phi = data[:, 0]     # First column, phi
psi = data[:, 1]     # Second column, psi

# Plotting the free energy surface
plt.scatter(phi, psi, c=-energy, cmap='viridis')
plt.colorbar(label='Negative Energy (kJ/mol)')
plt.xlabel('Phi Dihedral')
plt.ylabel('Psi Dihedral')
plt.title('Free Energy Surface')
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Figures/Lugano_exercise_1_A_plot_grid.png', dpi=300)
plt.show()
```
