```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md

# Load COLVAR data
colvar_file = '/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Setup/masterclass_exercise_1/A/COLVAR'
colvar_data = np.loadtxt(colvar_file, skiprows=1)  # Skip header
phi_colvar = colvar_data[:, 1]  # Assuming phi is in the second column
psi_colvar = colvar_data[:, 2]  # Assuming psi is in the third column

# Load trajectory
xtc_file = '/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Setup/masterclass_exercise_1/A/traj_comp.xtc'  # Change to your actual trajectory file
gro_file  = '/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Setup/masterclass_exercise_1/A/confout.gro'  # Change to your gro file
traj = md.load(xtc_file, top=gro_file)

# Compute phi and psi using MDTraj
phi_md, psi_md = md.compute_phi(traj), md.compute_psi(traj)
phi_md_angles = np.rad2deg(phi_md[0])  # Convert from radians to degrees
psi_md_angles = np.rad2deg(psi_md[0])  # Convert from radians to degrees

# Plotting
plt.figure(figsize=(10, 8))

# Scatter plot for phi vs psi
plt.scatter(phi_colvar, psi_colvar, color='blue', label='COLVAR data', alpha=0.5)
plt.scatter(phi_md_angles, psi_md_angles, color='red', label='MDTraj data', alpha=0.5)

plt.title('Ramachandran-like Plot')
plt.xlabel('Phi Dihedral (degrees)')
plt.ylabel('Psi Dihedral (degrees)')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='gray', lw=0.5, ls='--')
plt.axvline(0, color='gray', lw=0.5, ls='--')
plt.grid()
plt.legend()
# Save the figure in the Figures folder
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Figures/masterclass_exercise_1_A_plot_grid.png', dpi=300)
plt.show()

# Compute standard deviation
std_phi = np.std(phi_colvar)
std_psi = np.std(psi_colvar)

print(f'Standard deviation of phi: {std_phi:.2f} degrees')
print(f'Standard deviation of psi: {std_psi:.2f} degrees')
```
