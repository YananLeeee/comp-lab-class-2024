```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取自由能表面文件 (fes.dat)
fes_file = '/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Setup/Lugano_exercise/A/B10_S0.1_H1.2/B10_S0.1_H1.2_fes.dat'  # 请将此路径替换为您的 fes.dat 文件路径
fes_data_clean = pd.read_csv(fes_file, delim_whitespace=True, comment='#', header=None)

# 给数据添加列名
fes_data_clean.columns = ['phi', 'free_energy', 'der_phi']

# 绘制 phi vs free_energy 的自由能表面
plt.plot(fes_data_clean['phi'], fes_data_clean['free_energy'], label="Free Energy Surface")
plt.xlabel("Phi (radians)")
plt.ylabel("Free Energy (kJ/mol)")
plt.title("Free Energy Surface (FES)")
plt.legend()
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Figures/Lugano_exercise_2_A_plot_FES_Biasfactor10_Sigma0.1_Hillheight1.2.png', dpi=300)
# 显示图像
plt.show()
```
