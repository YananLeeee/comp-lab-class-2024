```python
import numpy as np
import matplotlib.pyplot as plt

# Load the FES data (assuming it's in a grid format)
fes_data = np.loadtxt('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Setup/2d_metadynamics/B10_S0.1&0.1_H1.2_10ns/fes.dat')

# Reshape if necessary (this depends on how your data is structured)
# Assume fes_data is in format [x, y, F]
x = fes_data[:, 0]
y = fes_data[:, 1]
F = fes_data[:, 2]

# Create a grid for phi, psi
X, Y = np.meshgrid(np.unique(x), np.unique(y))
F_grid = F.reshape(X.shape)

# Plot the free energy surface with contour lines
plt.figure()
contour_filled = plt.contourf(X, Y, F_grid, levels=20, cmap='viridis')
contours = plt.contour(X, Y, F_grid, colors='black', linewidths=0.5)
plt.clabel(contours, inline=True, fontsize=8)

# Link the colorbar to the filled contour plot
cbar = plt.colorbar(contour_filled)
cbar.set_label('Free Energy (kT)')

plt.xlabel('phi (degrees)')
plt.ylabel('psi (degrees)')
plt.title('Free Energy Surface from 2D Metadynamics')
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week6-Metadynamics/Figures/2d_metadynamics_A_plot_2D_FES_Biasfactor10_Sigma0.1_Hillheight1.2_10ns.png', dpi=300)
plt.show()
```
