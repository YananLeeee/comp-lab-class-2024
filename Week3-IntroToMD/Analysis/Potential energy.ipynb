```python
import pandas as pd
import matplotlib.pyplot as plt

# Initialize a list to hold the data
data = []

# Read the potential.xvg file manually
with open('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Data/potential.xvg', 'r') as f:
    for line in f:
        if not line.startswith(('@', '#')):  # Skip comments
            fields = line.split()
            if len(fields) == 2:  # Adjust based on your data structure
                try:
                    data.append([float(fields[0]), float(fields[1])])
                except ValueError:
                    print(f"Skipping line due to conversion error: {line}")

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Step', 'Potential Energy'])

# Check the first few rows
print(df.head())

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Step'], df['Potential Energy'], label='Potential Energy', color='b', linewidth=3)
plt.title('Energy Minimization\n2JOF,Steepest Descent',fontsize=26)
plt.xlabel('Energy Minimization Step',fontsize=26)
plt.ylabel('Potential Energy (KJ mol⁻¹)',fontsize=26)
plt.tick_params(axis='both', labelsize=22)
plt.grid()
plt.legend(fontsize=24)
plt.gca().spines['top'].set_linewidth(3)    # 上边框
plt.gca().spines['right'].set_linewidth(3)  # 右边框
plt.gca().spines['left'].set_linewidth(3)   # 左边框
plt.gca().spines['bottom'].set_linewidth(3) # 下边框
# 保存图像到文件
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Figures/potential_plot.png', dpi=300, bbox_inches='tight')  # 保存为PNG文件，300 DPI清晰度
plt.show()
```
