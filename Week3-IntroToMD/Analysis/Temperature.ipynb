```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to the temperature.xvg file
file_path = '/home/yl12451/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Data/temperature.xvg'

# Initialize a list to hold the data
data = []

# Read the file manually
with open(file_path, 'r') as f:
    for line in f:
        if not line.startswith(('@', '#')):  # Skip comments
            fields = line.split()
            if len(fields) == 2:  # Adjust based on expected columns
                try:
                    data.append([float(fields[0]), float(fields[1])])
                except ValueError:
                    print(f"Skipping line due to conversion error: {line}")

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Time (ps)', 'Temperature'])

# Check the first few rows
print(df.head())

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Time (ps)'], df['Temperature'], label='Temperature', color='b',linewidth=3)
plt.title('Temperature\n2JOF, NVT, Equilibration', fontsize=26)
plt.xlabel('Time (ps)', fontsize=26)
plt.ylabel('Temperature (K)', fontsize=26)
plt.tick_params(axis='both', labelsize=22)
plt.grid()
plt.legend(fontsize=24)
plt.gca().spines['top'].set_linewidth(3)    # 
plt.gca().spines['right'].set_linewidth(3)  # 
plt.gca().spines['left'].set_linewidth(3)   # 
plt.gca().spines['bottom'].set_linewidth(3) # 
# 
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Figures/temperature_plot.png', dpi=300, bbox_inches='tight')  # PNG300 DPI
plt.show()
```
