```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to the pressure.xvg file
file_path = '/home/yl12451/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Data/pressure.xvg'

# Initialize a list to hold the data
data = []

# Read the pressure.xvg file manually
with open(file_path, 'r') as f:
    for line in f:
        if not line.startswith(('@', '#')):  # Skip comment lines
            fields = line.split()
            if len(fields) == 2:  # Adjust based on expected columns
                try:
                    data.append([float(fields[0]), float(fields[1])])
                except ValueError:
                    print(f"Skipping line due to conversion error: {line}")

# Convert the data to a DataFrame
df = pd.DataFrame(data, columns=['Time (ps)', 'Pressure (bar)'])

# Check the first few rows of the DataFrame
print(df.head())

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Time (ps)'], df['Pressure (bar)'], label='10-ps Running Avg', color='b', linewidth=3)

# Add title and labels
plt.title('Pressure\n2JOF, NPT Equilibration', fontsize=26)
plt.xlabel('Time (ps)', fontsize=26)
plt.ylabel('Pressure (bar)', fontsize=26)

# Set tick parameters for better readability
plt.tick_params(axis='both', labelsize=22)

# Enable grid and legend
plt.grid()
plt.legend(fontsize=24)

# Adjusting spine widths for a clearer border
for spine in plt.gca().spines.values():
    spine.set_linewidth(3)

# 保存图像到文件
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Figures/pressure_plot.png', dpi=300, bbox_inches='tight')  # 保存为PNG文件，300 DPI清晰度

# Show the plot
plt.show()
```
