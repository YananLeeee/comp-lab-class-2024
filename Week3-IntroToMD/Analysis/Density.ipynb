```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to the density.xvg file
file_path = '/home/yl12451/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Data/density.xvg'

# Initialize a list to hold the data
data = []

# Read the density.xvg file manually
with open(file_path, 'r') as f:
    for line in f:
        if not line.startswith(('@', '#')):  # Skip comment lines
            fields = line.split()
            if len(fields) == 2:  # Adjust based on expected columns
                try:
                    data.append([float(fields[0]), float(fields[1])])
                except ValueError:
                    print(f"Skipping line due to conversion error: {line}")

# Convert the data to a DataFrame
df = pd.DataFrame(data, columns=['Time (ps)', 'Density (kg m⁻³)'])

# Check the first few rows of the DataFrame
print(df.head())

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Time (ps)'], df['Density (kg m⁻³)'], label='10-ps Running Avg', color='b', linewidth=3)
plt.text(0.5, 1.05, 'Density', fontsize=26, ha='center', va='bottom')  # 主标题
plt.text(0.5, 1.00, ' 2JOF,NPT Equilibration', fontsize=20, ha='center', va='bottom')  # 子标题

# Add title and labels 
#plt.title('Density\n2JOF, NPT Equilibration', fontsize=26)
plt.xlabel('Time in picoseconds (ps)', fontsize=26)
plt.ylabel('Density in kg m⁻³', fontsize=26)

# Set tick parameters for better readability
plt.tick_params(axis='both', labelsize=22)

# Enable grid and legend
plt.grid()
plt.legend(fontsize=24)

# Adjusting spine widths for a clearer border
for spine in plt.gca().spines.values():
    spine.set_linewidth(3)

# 保存图像到文件
plt.savefig('/home/yl12451/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Figures/density_plot.png', dpi=300, bbox_inches='tight')  # 保存为PNG文件，300 DPI清晰度
# 显示图表
plt.show()
```
