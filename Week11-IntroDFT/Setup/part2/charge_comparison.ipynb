```python
from psikit import Psikit
```

```python
pk = Psikit()
```

```python
pk.read_from_smiles('CN1C=NC2=C1C(=O)N(C(=O)N2C)C')
```

```python
pk.optimize()
```

```python
pk.mol
```

```python
#caculate chages
print(pk.calc_resp_charges())
print(pk.calc_mulliken_charges())
print(pk.calc_lowdin_charges())
```

```python
mol = pk.mol
atoms = mol.GetAtoms()
```

```python
data = {'Symbol':[atom.GetSymbol() for atom in atoms],
        'RESP': [float(atom.GetProp('RESP')) for atom in atoms],
        'MULLIKEN': [float(atom.GetProp('MULLIKEN')) for atom in atoms],
       'LOWDIN':  [float(atom.GetProp('LOWDIN')) for atom in atoms]}
```

```python
df = pd.DataFrame(data)
```

```python
df.round(3)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
# RESP charge and MULLIKEN charge shows high correlation
plt.scatter(df.RESP, df.MULLIKEN)
plt.xlabel('RESP')
plt.ylabel('MULLIKEN')
plt.plot(np.arange(-1.0, 1.0, 0.01), np.arange(-1.0, 1.0, 0.01), c='b')
```

```python
plt.scatter(df.RESP, df.LOWDIN)
plt.xlabel('RESP')
plt.ylabel('LOWDIN')
plt.plot(np.arange(-1.0, 1.0, 0.01), np.arange(-1.0, 1.0, 0.01), c='b')
```

```python
plt.scatter(df.MULLIKEN, df.LOWDIN)
plt.xlabel('MULLIKEN')
plt.ylabel('LOWDIN')
plt.plot(np.arange(-1.0, 1.0, 0.01), np.arange(-1.0, 1.0, 0.01), c='b')
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Get RESP charges
resp_charges = pk.calc_resp_charges()

# Create a copy of the molecule to avoid modifying the original
mol_with_charges = Chem.Mol(pk.mol)

# Set custom labels for each atom with RESP charges
for i, atom in enumerate(mol_with_charges.GetAtoms()):
    atom.SetProp("atomLabel", f"{resp_charges[i]:.2f}")  # Use 2 decimal places for readability

# Draw the molecule with custom atom labels
img = Draw.MolToImage(mol_with_charges, size=(300, 300), kekulize=True)

# Display the image in the Jupyter notebook
from IPython.display import display
display(img)
```
